```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
urOn = 'https://movies.yahoo.com.tw/movie_intheaters.html'
urlNext ="https://movies.yahoo.com.tw/movie_comingsoon.html"
movies = []
all=["url","ur0n","urlNext"]

def crawl(url):

  r = requests.get(url)
  yahoo = BeautifulSoup(r.text, 'html5lib')

  movie_infos = yahoo.find_all('div', class_="release_info_text")
  for movie_info in movie_infos:


    name = movie_info.find('div', class_="release_movie_name").find('a',
    class_="gabtn").get_text(strip=True)

    en_name = movie_info.find('div', class_="en").find('a',
    class_="gabtn").get_text(strip=True)

    movie_date = movie_info.find('div', class_="release_movie_time").get_text(strip=True)[5:]

    exception = movie_info.find('div', class_ = "leveltext").find('span').get_text(strip=True)
    movies.append([name, en_name, movie_date, exception])

    df = pd.DataFrame(movies,columns=['中文名稱','英文名稱','上映日期','期待度'])
  print(df)
crawl(url)
crawl(urOn)
crawl(urlNext)
```
